```python
def direct_query_with_pandas(question, df):
    """
    Fonction pour filtrer directement avec pandas et générer une réponse
    
    Args:
        question (str): Question en langage naturel
        df (DataFrame): DataFrame des bâtiments
        
    Returns:
        str: Réponse en langage naturel
    """
    # Normalisation de la question (suppression des accents, minuscules)
    import unicodedata
    question_norm = unicodedata.normalize('NFKD', question.lower()).encode('ASCII', 'ignore').decode('utf-8')
    
    # Question 1: "Quels bâtiments résidentiels de plus de 1000 m² sont classés F ou G dans le département 93 ?"
    if all(term in question_norm for term in ["batiment", "resident", "1000", "class", "f ou g", "93"]):
        # Filtrage direct avec pandas
        result_df = df[(df['est_residence'] == True) & 
                     (df['surface_totale'] > 1000) & 
                     (df['classe_dpe'].isin(['F', 'G'])) & 
                     (df['departement'] == '93')]
        
        # Génération d'une réponse
        response = f"J'ai trouvé {len(result_df)} bâtiments résidentiels de plus de 1000 m² classés F ou G dans le département 93.\n\n"
        
        if len(result_df) > 0:
            response += "Voici les 5 premiers:\n"
            for idx, row in result_df.head(5).iterrows():
                response += f"- Bâtiment ID {row['id_bdnb']} à {row.get('commune', 'commune non renseignée')}, "
                response += f"surface de {row['surface_totale']:.1f} m², classe DPE {row['classe_dpe']}\n"
            
            if len(result_df) > 5:
                response += f"... et {len(result_df) - 5} autres bâtiments."
        
        return response
    
    # Question 2: "Quelle est la surface moyenne des bâtiments tertiaires construits avant 1975 dans le département du Rhône ?"
    elif all(term in question_norm for term in ["surface", "moyenne", "tertiaire", "1975", "rhone"]):
        # Identification du département du Rhône (69)
        dept_rhone = '69'
        
        # Filtrage
        result_df = df[(df['est_tertiaire'] == True) & 
                       (df['annee_construction'] < 1975) & 
                       (df['annee_construction'] > 0) &  # Exclure les valeurs non renseignées
                       ((df['departement'] == dept_rhone) | 
                        (df['departement'].str.startswith('69') if pd.api.types.is_string_dtype(df['departement']) else False))]
        
        if len(result_df) > 0:
            surface_moyenne = result_df['surface_totale'].mean()
            nb_batiments = len(result_df)
            response = f"La surface moyenne des {nb_batiments} bâtiments tertiaires construits avant 1975 dans le département du Rhône est de {surface_moyenne:.2f} m²."
        else:
            response = "Je n'ai pas trouvé de bâtiments tertiaires construits avant 1975 dans le département du Rhône avec des données de surface valides."
            
        return response
    
    # Question 3: "Quels sont les 10 quartiers de Marseille avec le plus de passoires thermiques ?"
    elif all(term in question_norm for term in ["quartier", "marseille", "passoire"]):
        # Filtrage pour Marseille
        marseille_mask = ((df['commune'] == 'MARSEILLE') | 
                         (df['commune'].str.contains('MARSEILLE', case=False, na=False) if pd.api.types.is_string_dtype(df['commune']) else False) |
                         (df['adr_commune'] == 'MARSEILLE') | 
                         (df['adr_commune'].str.contains('MARSEILLE', case=False, na=False) if pd.api.types.is_string_dtype(df['adr_commune']) else False))
        
        marseille_df = df[marseille_mask]
        
        # Vérification que nous avons des données de quartier
        if 'quartier' in marseille_df.columns and marseille_df['quartier'].notna().any():
            quartier_column = 'quartier'
        elif 'arrondissement' in marseille_df.columns and marseille_df['arrondissement'].notna().any():
            quartier_column = 'arrondissement'
        else:
            return "Je ne dispose pas d'informations sur les quartiers de Marseille dans ce jeu de données."
        
        # Comptage des passoires thermiques par quartier
        passoires_par_quartier = marseille_df[marseille_df['est_passoire_thermique'] == True].groupby(quartier_column).size()
        passoires_par_quartier = passoires_par_quartier.sort_values(ascending=False).head(10)
        
        if len(passoires_par_quartier) > 0:
            response = "Les 10 quartiers de Marseille avec le plus de passoires thermiques sont:\n\n"
            for quartier, count in passoires_par_quartier.items():
                response += f"- {quartier}: {count} passoires thermiques\n"
        else:
            response = "Je n'ai pas trouvé d'informations sur les passoires thermiques dans les quartiers de Marseille."
            
        return response
    
    # Question 4: "Quel est le pourcentage de bâtiments résidentiels construits avant 1948 à Lyon ?"
    elif all(term in question_norm for term in ["pourcentage", "resident", "1948", "lyon"]):
        # Filtrage pour Lyon
        lyon_mask = ((df['commune'] == 'LYON') | 
                    (df['commune'].str.contains('LYON', case=False, na=False) if pd.api.types.is_string_dtype(df['commune']) else False) |
                    (df['adr_commune'] == 'LYON') | 
                    (df['adr_commune'].str.contains('LYON', case=False, na=False) if pd.api.types.is_string_dtype(df['adr_commune']) else False))
        
        lyon_df = df[lyon_mask & (df['est_residence'] == True)]
        
        if len(lyon_df) > 0:
            # Nombre total de bâtiments résidentiels à Lyon
            total_residentiels = len(lyon_df)
            
            # Nombre de bâtiments construits avant 1948
            construits_avant_1948 = len(lyon_df[(lyon_df['annee_construction'] < 1948) & (lyon_df['annee_construction'] > 0)])
            
            # Calcul du pourcentage
            if total_residentiels > 0:
                pourcentage = (construits_avant_1948 / total_residentiels) * 100
                response = f"À Lyon, {pourcentage:.1f}% des bâtiments résidentiels ont été construits avant 1948 ({construits_avant_1948} sur un total de {total_residentiels})."
            else:
                response = "Je n'ai pas trouvé de bâtiments résidentiels à Lyon dans le jeu de données."
        else:
            response = "Je n'ai pas trouvé de bâtiments résidentiels à Lyon dans le jeu de données."
            
        return response
    
    # Question 5: "Quels sont les bâtiments avec plus de 5 étages dans le 6e arrondissement de Paris ?"
    elif all(term in question_norm for term in ["batiment", "etage", "arrondissement", "paris"]):
        # Extract "6e" or other ordinal indicators
        import re
        arr_match = re.search(r'(\d+)[eè]?me?\s+arrondissement', question_norm)
        if arr_match:
            arr_num = arr_match.group(1)
            
            # Filtrage pour Paris 6e
            paris_6e_mask = (
                ((df['commune'] == 'PARIS') | 
                (df['commune'].str.contains('PARIS', case=False, na=False) if pd.api.types.is_string_dtype(df['commune']) else False) |
                (df['adr_commune'] == 'PARIS') | 
                (df['adr_commune'].str.contains('PARIS', case=False, na=False) if pd.api.types.is_string_dtype(df['adr_commune']) else False)) &
                ((df['arrondissement'] == arr_num) | 
                (df['arrondissement'] == f"PARIS {arr_num}E") |
                (df['arrondissement'] == f"PARIS {arr_num}EME") |
                (df['arrondissement'] == f"75{arr_num.zfill(2)}") |
                (df['arrondissement'].astype(str).str.contains(f"PARIS.{arr_num}", case=False, na=False, regex=True) if pd.api.types.is_string_dtype(df['arrondissement']) else False))
            )
            
            result_df = df[paris_6e_mask & (df['nb_etages'] > 5)]
            
            if len(result_df) > 0:
                response = f"J'ai trouvé {len(result_df)} bâtiments avec plus de 5 étages dans le {arr_num}e arrondissement de Paris.\n\n"
                response += "Voici les 10 premiers:\n"
                
                for idx, row in result_df.head(10).iterrows():
                    response += f"- Bâtiment ID {row['id_bdnb']} à {row.get('adresse', 'adresse non renseignée')}, "
                    response += f"avec {int(row['nb_etages'])} étages"
                    if pd.notna(row['type_batiment']):
                        response += f", type: {row['type_batiment']}"
                    response += "\n"
                
                if len(result_df) > 10:
                    response += f"... et {len(result_df) - 10} autres bâtiments."
            else:
                response = f"Je n'ai pas trouvé de bâtiments avec plus de 5 étages dans le {arr_num}e arrondissement de Paris."
                
            return response
    
    # Si aucun pattern ne correspond, retourner None pour utiliser le RAG
    return None
    
    # Ajouter d'autres patterns de questions ici
    
    # Si aucun pattern ne correspond, utiliser le RAG
    return None

# Fonction combinée
def intelligent_query(question, df):
    """
    Choisit entre la recherche directe pandas ou le RAG en fonction de la question
    """
    direct_response = direct_query_with_pandas(question, df)
    if direct_response:
        return direct_response
    else:
        return query_assistant(question)["answer"]
```
